```python
x=dataset.iloc[:,[0,18]]
x1=x.iloc[:,[1]]
x1
```

```python
from sklearn.cluster import KMeans
```

```python
def elbow_method(x):
    wcss=[]
    for i in range(1,11):
        kmeans=KMeans(i,init='k-means++',max_iter=300,n_init=10,random_state=0)
        kmeans.fit(x)
        wcss.append(kmeans.inertia_)
    plt.plot(range(1,11),wcss)
    plt.xlabel('No. of clusters')
    plt.ylabel('wcss')
    plt.title('elbow method')
    plt.show()
```

```python
elbow_method(x1)
```

```python
def k_means(n,x):
    kmeans=KMeans(n,init='k-means++',max_iter=300,n_init=10,random_state=0)
    y_kmeans=kmeans.fit_predict(x)
    cluster_centers=kmeans.cluster_centers_
    return y_kmeans,cluster_centers
```

```python
labels,cluster_centers=k_means(4,x1)
print(labels)
print(cluster_centers)
```

```python
print((list(labels)))
```

```python
x['Student_Density']=list(labels)
x
```

```python
x.drop(['cluster'],axis=1)
```

```python
stud_density=x[['college_institution_id','Student_Density']]
stud_density
'''
0-very low
1-low
2-medium
3- high
'''
```

```python
df=pd.merge(dataset,stud_density,on='college_institution_id',how='inner')
df
```
