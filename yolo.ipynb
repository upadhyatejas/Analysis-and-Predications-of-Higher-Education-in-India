```python
x = accr2.iloc[:,[0,8,53]]
x1 = x.iloc[:,[2]]
x1
```

```python
from sklearn.cluster import KMeans
```

```python
def elbow_method(x):
    wcss=[]
    for i in range(1,11):
        kmeans=KMeans(i,init='k-means++',max_iter=300,n_init=10,random_state=0)
        kmeans.fit(x)
        wcss.append(kmeans.inertia_)
    plt.plot(range(1,11),wcss)
    plt.xlabel('No. of clusters')
    plt.ylabel('wcss')
    plt.title('elbow method')
    plt.show()
```

```python
elbow_method(x1)
```

```python
def k_means(n,x):
    kmeans=KMeans(n,init='k-means++',max_iter=300,n_init=10,random_state=0)
    y_kmeans=kmeans.fit_predict(x)
    cluster_centers=kmeans.cluster_centers_
    return y_kmeans,cluster_centers
```

```python
# I have used 5 labels (excellent,very good,good,satisfactory,bad) and 5 clusters results in better distribution
labels,cluster_centers=k_means(5,x1)
print(labels)
print(cluster_centers)
```

```python
print(np.array(labels))
```

```python
x['infrastructure_class']=list(labels)
x
```

```python
infraclass = x[["college_institution_id","infrastructure_class"]]
```
